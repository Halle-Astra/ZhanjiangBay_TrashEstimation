# 用于论文绘图

假设海岸线

$y = 0.1(x-5)^2-5$

第一个水流的碰撞点假设为$(-3,1.4)$（将-3代入海岸线方程）

海岸线方程导数为

$y=0.2(x-5)$

第一个点的导数为1.6，法线斜率为1/1.6....剩下的我下床手算好了。

```python
from matplotlib import pyplot as plt 
import numpy as np 
%matplotlib inline 

def plot_line(a,b,style = '-'):
    plt.plot([a[0],b[0]],[a[1],b[1]],'k'+style)

reflect2 = lambda x:-0.3*x+0.5#AD
x_reflect2 = lambda x:(x-0.5)/(-0.3)
x1 = np.arange(-4,5,0.1)
y1 = 0.1*(x1-5)**2-5
y_y1 = lambda x:0.1*(x-5)**2-5
x_coast = lambda y:((y+5)*10)**0.5+5

S = np.array([-2.5,2.7])
C = np.array([0,reflect2(0)])
A = np.array([-3,1.4])
B = np.array([-1,y_y1(-1)])
D = np.array([x_reflect2(B[1]),B[1]])
E = np.array([C[0],D[1]])
F = np.array([C[0],A[1]])
AC_ar = 0.5*(A+C)
BC_ar = 0.5*(B+C)

plt.plot(x1,y1,'k')
plot_line(A,C)
plot_line(B,C)
plt.annotate('',AC_ar,A,arrowprops=dict(arrowstyle = '->',linewidth = 2 ))
plt.annotate('',BC_ar,B,arrowprops=dict(arrowstyle = '->',linewidth = 2))
plt.annotate('',A,S,arrowprops=dict(arrowstyle = '->',linewidth = 2))
plt.hlines(B[1],-4,7,linestyle = 'dashed')
plt.hlines(A[1],-4,4,linestyle = 'dashed')
plot_line(C,D,'--')
plot_line(E,F,'--')
plt.text(A[0]-0.5,A[1]-0.5,'A')
plt.text(B[0]-0.5,B[1]-0.5,'B')
plt.text(C[0]+0.2,C[1],'C')
plt.text(D[0],D[1],'D')
plt.text(E[0],E[1]-0.5,'E')
plt.text(F[0],F[1]+0.2,'F')
plt.text(S[0],S[1],'S')
plt.axis('equal')
plt.grid('on')
plt.axis('on')
#plt.box('off')
#plt.xticks([])
#plt.yticks([])
plt.axis('off')
plt.ylim(-2,2)
plt.savefig('cross_point.png',dpi = 1000)
plt.show()
```
